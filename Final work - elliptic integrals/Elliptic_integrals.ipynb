#**<font color=#6600cc>Introduction</font>**

Our problem is finding a way to solve the integral 

$I =\int_{1}^{\infty} \frac{dx}{\sqrt{x^{4}-1}}\ \ \ $   (A)

However, the integral cannot be evaluate in terms of elementals functions, then we do a series of changes of variable as follows

First, if  $x = \dfrac{1}{t}$, then $dx = -\dfrac{dt}{t^{2}}$, and when $x = 1, t = 1$ and when $x = \infty, t = 0$ 

the (A) is transformed to 

$I =\int_{0}^{1} \frac{dt}{\sqrt{1-t^{4}}}\ \ \ $   (B)

second, let $t = sin\phi$, $dt = cos\phi\ d\phi$, when $t = 0, \theta = 0$ and when $t = 1, \theta = \pi/2$, giving

$I =\int_{0}^{\pi/2} \frac{d\phi}{\sqrt{2-cos^{2}\phi}}\ \ \ $   (C)

Finally, let $\phi=\pi/2-\theta$ to get


$I =\dfrac{1}{\sqrt{2}} \int_{0}^{\pi/2} \frac{d\theta}{\sqrt{1-\dfrac{sin^{2}\theta}{2}} }\ \ \ $   (D)

We have an integral  of **elliptic integral form** then

$I =\int_{0}^{1} \frac{dt}{\sqrt{1-t^{4}}} = \dfrac{1}{\sqrt{2}} \int_{0}^{\pi/2} \frac{d\theta}{\sqrt{1-\dfrac{sin^{2}\theta}{2}} } $







#**<font color=#333399>Section 1: Elliptic integrals</font>**

#**1.1 Definition** 
The *complete elliptic integral of first kind*, denoted by $K(m)$, $K$, or $K(k)$ (with $m=k^{2}$), is

$
K(m)=K(k)=\int_{0}^{\pi/2} \frac{d\theta}{\sqrt{1-k^{2} sin^{2}\theta}} \ \  0 < k < 1  \ \ \ (1.1)
$ 

while the *incomplete elliptic integral of the first kind*, denoted by $K(\phi|k)$ or $k(\phi|m)$ is

$
K(\phi|m)=K(\phi|k)=\int_{0}^{\phi} \frac{d\theta}{\sqrt{1-k^{2} sin^{2}\theta}}   \ \ \ (1.2)
$ 


with $0 \leq \phi \leq \pi/2$  


While the *complete elliptic integral of second kind*, denoted by $E(m)$, $K$, or (with $m=k^{2}$), is

$
E(k)=\int_{0}^{\pi/2}  \sqrt{1-k^{2} sin^{2}\theta} d\theta, \ \ \ 0 < k < 1
$ 

the *incomplete elliptic integral of the second kind*, denoted by $E(\phi|k)$ or $E(\phi|m)$ is

$
E(\phi|m)=E(\phi|k)=\int_{0}^{\phi}  \sqrt{1-k^{2} sin^{2}\theta} d\theta   \ \ \ 
$ 


with $0 \leq \phi \leq \pi/2$  


Finally the incomplete integral of third kind is 

$
H(\phi,n,k)=\int_{0}^{\phi} \frac{d\theta}{(1+n sin^{2}\theta)\sqrt{1-k^{2} sin^{2}\theta}}   \ \ \ 0 < k  < 1 \ \  
$ 






# **1.2 Series of elliptic integrals of first kind**





$K(k) = \int_{0}^{\pi/2} \frac{d\theta}{\sqrt{1-k^{2} sin^{2}\theta}} =\sum_{n=0}^\infty \int_{0}^{\pi/2} \dfrac{(2n)!}{2^{2n}  (n!)^{2}} k^{2n} sin^{2n} \theta\ d\theta  \ \ \ (1.3) $

By Wallis' formula, we get

$K =\sum_{n=0}^\infty  \left[\dfrac{(2n)!}{2^{2n}(n!)^{2}}  \right]^{2}k^{2n}  \ \ \ (1.4)  $

According to Newton´s law, where the force per unit mas is $\frac{\mu}{r^2}$, a planet moves around the Sun in a ellipse remains the same forever, according to Newton´s law, the curvature of space appears as a correction term proportional to $\frac{1}{r^{4}}$, the gravitational force per unit of mass is of the form 

$\mu(\frac{1}{r^{2}}+\frac{3h^{2}}{c^{2}r^{4}}) \ \ \ (2.1)$

$h$  is the angular momentum per unit massof the planet about the sun, $c$ is the speed of light. The prediction of Einstein is the path is very nearly  an ellipse, but it does not quite close up, in the next revolution, the path will have advnced slightly ahead in which the planet is moving, then the orbit is an ellipse which very slowly precesses.

# **2.2 Arc length of an ellipse**
The equation of the ellipse 

$\frac{x^{2}}{a^{2}}+\frac{y^{2}}{b^{2}} = 1$ 

Parametrisation of an arc of the ellipse
$ (x(\phi), y(\phi)) = (a sin \phi, b cos \phi ), \ \ \ \phi \in [0,\theta] $

the length of the arc is 


$\int_{0}^{\theta} \sqrt{(\frac{d}{d\theta} a sin\phi)^{2} + (\frac{d}{d\theta} a sin\phi)^{2}} d\phi = a \int_{0}^{\theta} \sqrt{1-k^{2} sin^{2} (\phi)} d\phi$

where $k = \sqrt{(a^{2}-b^{2})/a^{2}}$

then we have an incomplete elliptic integral of the second kind: 

$ E(k,\theta) = \int_{0}^{\theta} \sqrt{1-k^{2} sin^{2} (\phi)} d\phi$

# **4.1 Numerical Integration**

#**4.1.1 Simpson´s rule**

Simpson's rule uses a quadratic polynomial on each subinterval of a partition to approximate the function  and to compute the definite integral. This is an improvement over the trapezoid rule which approximates  by a straight line on each subinterval of a partition.

![image 2](https://drive.google.com/uc?export=view&id=16Jb5QBTWx-9560cpwlwjzWMYFWIT3HIT)






The formula for Simpson's rule is


$S_N(f) = \frac{\Delta x}{3} \sum_{i=1}^{N/2} \left( f(x_{2i-2}) + 4 f(x_{2i-1}) + f(x_{2i}) \right)\ \ \ \ $     (4.1)


$N$ is a even number of subintervals $[a,b]$, $\Delta x = (b - a)/N$ and $x_i = a + i \Delta x \ \ \ \ $ (4.2) 


We can re-write Simpson´s rule by gropuing as follows 

$\int_a^bf(x)dx \approx \frac{(\Delta x)}{3} [y_0+4(y_1+y_3+y_5+...) + 2(y_2+y_4+y_6+...)+y_n] \ \ \ \  $ (4.3)

Then

$\int_a^bf(x)dx \approx \frac{(\Delta x)}{3} [First \ term +4(Sum\  of \ odd \ terms) + 2(sum\ of\ even\ terms)+last] \ \ \ \  (4.4)$


#**4.1.2 Clenshaw-Curtis aproximation**

This method approaches the integral 

$ I = \int_{-1}^{1} f(x)dx$

by sums

$I_{n} = \sum_{k=0}^{n} \omega_{k} f(x_{k}) \ \ \ (4.5)$ 

for various $n$, where the nodes depend on $n$ but not  $f$. The set {$\omega_{k}$} are defined only by the property that $I_{n}$ is equal to the integral of the degree $\leq n$.

The method of Clensaw-Curtis quadrature is to use Chebysev points instead of optimal nodes. The nodes, Chebysev points, are edfined by 

$x_{j} = cos(\frac{j\pi}{n}), \ \ \ 0 \leq j \leq n $


The Fourier series of general function $f(x)$

$f(x) = a_{0} + \sum_{n=1}^{\infty} a_{n}cos(nx)+\sum_{n=1}^{\infty} b_{n}sin(nx)$

with only the constant and the cosine terms is knwon as a "Fourier cosine series", a Chebyshev series is a Fourier cosine series with a change of variable

$x = cos(\theta)$

and then 

$T_{n}(x) = cos(n\theta)$

The following series are then equivalent under transformation 


$f(x) = \sum_{n=0}^{\infty} a_{n}T_{n}(x) $


$f(cos\theta) = \sum_{n=0}^{\infty} a_{n}cos(n\theta) $

This can be seen when   $x = cos(\theta)$ and $d/dx=(-1/sin\theta)(d/d\theta)$ the diferential equation that satisfies the type I Chebysev polynomial  

$(1-x^2)T_{n}''(x)-xT_{n}'(x)+n^2T_{n}(x)=0$


to 

$ \dfrac{d^{2} T_{n}}{d\theta^{2}} +n^{2}T_{n}=0$

the simple harmonic oscillator equation with solutions $cos (n\theta)$ and $sin(n\theta)$




#**4.2 Small value aproximation**

Develop the integrand of 

$\int_{0}^{\pi/2} \frac{d\theta}{\sqrt{1-k^{2} sin^{2}\theta}}  $

using binomial expansion

$(1+b)^{n} = 1 + nb + \frac{n(n-1)}{2!}b^{2} + \frac{n(n-1)(n-2)b^{3}}{3!} \ \ \ (4.6)$

for small values of $m$


$\int_{0}^{\pi/2} \frac{d\theta}{\sqrt{1-k^{2} sin^{2}\theta}}  = \int_{0}^{\pi/2} (1-k^{2} sin^{2}\theta)^{-1/2}  d\theta \approx \int_{0}^{\pi/2} (1+\frac{m sin^{2}\theta}{2})  d\theta = \frac{\pi}{2}(1+\frac{m}{4}) \ \ \ (4.7) $

#**<font color=#333399>Section 5: Implementation of algorithm to solve Elliptic integral with numerical integration</font>**

Libraries

```
%pylab inline
from scipy import integrate
import math 
import pandas as pd
import time
import matplotlib.pyplot as plt
```

Functions

```
def simpson_function(f, a, b, N, args=None): 
  """
  This function implements the Simpson´s rule, giving by eq. (4),  needs four inputs:
  f: The function that we´ll integrate
  a,b: The start point and the end point of interval of integration [a,b]
  N: The number of equal segments of width h
  arg: list of arguments for the function
  """
  if args is None:
   args = []                        # If the arguments of the function is none, then create a list empty
  x=linspace(a,b,2*N)               # Create a vector  with 2N items between a and b
  h=(b-a)/(2*N-1)                   # Crete the length of segments, for this substract the start point and the start point
                                    # and then divided by space for 2N elements
  y=f(*[x] + list(args))            # Apply the function to x (in form of list)
                                    # If the number of args is unknown, add * before the parameter name
                                    # Slice function in list: Notation [start:stop:step]
  evens = y[2:-2:2]                 # Take the items from index=2 to index = -2 in steps of 3 items 
  odds= y[1:-1:2]                   #
  return (y[0] + 2*odds.sum()+ 4*evens.sum() + y[-1] )*h/3.0  # Do the sum 




def K_int(phi,m):                   # Create the function to integrate 
  return 1.0/sqrt(1-m*sin(phi)**2)  # The function is giving by (1.)



def K(m,N):                                  # Apply simpson function to each value of m
  return simpson_function(K_int, 0, pi/2, N=N,args=(m,))

def series(n,k):    # k = sqrt(m)
  SeriesValue = []   
  for j in range(n): # 0 to n-1
    z  = (math.factorial(2*j))/((2**(2*j))*(math.factorial(j))**2)
    zsquare= z**2
    z1 =  zsquare*(k**(2*j))*pi/2
    SeriesValue.append(z1)
  return SeriesValue
```

Operations

```
# Value specific 

thetaspecific = 90                    # Angle (degrees) to calculate      
thetaRad     = thetaspecific*pi/180   # Convert to radians
m=sin(thetaRad/2)**2                  # get the value of m in this case m=1/2

simpsresult = K(m,1000)                    # Apply the simpson aproximation to integral ()

quadResult = integrate.quad(K_int, 0, pi/2, args=(m,))      # Apply the Clenshaw–Curtis quadrature to integral ()

smallAngle = (pi/2)*(1.0+m/4)         #   Approxmation for small angle in the series to integral ()

SeriesResult = series(1000,1/sqrt(2)) # Aproximation for series to integral ()
SeriesValuet=sum(SeriesResult) 


value = 1/4                           # Analytical result to integral ()
gammavalue = math.gamma(value)
gamaresult = ((gammavalue)**2)/(4*sqrt(2*pi))
```

Preprocessing of results

```
SimpsonValue    = simpsresult/sqrt(2) 
ClenshaowValue  = quadResult/sqrt(2)     
SmallAngleValue = smallAngle/sqrt(2)   
SeriesValNorm   = SeriesValuet/sqrt(2)          
```

Comparison

```
Values = {'Method': ['Simpson´s Rule','Clenshaow ','Analytic','Series','Small Angle'],
        'Value': [SimpsonValue ,ClenshaowValue[0],gamaresult,SeriesValNorm,SmallAngleValue]
        }
df = pd.DataFrame(Values, columns = ['Method', 'Value'])
print (df)
```

# Calculation for several $m$ values

```
## We create a empty lists, where the results will be stored

SimpsonList=[]                                  #   Store the results from numeric integation done by Simpson´s rule
QuadList=[]                                     #   Store the results from numeric integration done whith Clenshaw–Curtis quadrature
                                                #   that are based on an expansion of the integrand in terms of Chebyshev polynomials
                                                #   this thecnique is obtained from library scipy
SmallAngleList=[]                               #   Store the results from approximation of small angle 

SeriesList = []                                 #   Series

# Create a set of 150 angles between 0° and 90°  
thetdeg = linspace(0.0, 90, 150)

# Cicle for each angle of 150 angles in thetArray
for theta in thetdeg:
  thetaRad = theta*pi/180                       #   Convert the angles in degrees to radians
  m=sin(thetaRad/2)**2                          #   Apply equivalence to m 
  simpsresult = K(m,1000)                            #   Calculate 
  quadResult = integrate.quad(K_int, 0, pi/2, args=(m,))      # Parmeters of integrate.quad 
                                                #   integrate.quad(Function,lower limit of integration,...
                                                #   upper limit of integration, extra arguments of the function)
  smallAngle = (pi/2)*(1.0+m/4)                 #   Approxmation for small angle in the series
  
  seriesval = sum(series(1000,sqrt(m)))

                                                #   Finally we append the result normalized for each angle in the empty list  
  SimpsonList.append(2*simpsresult/pi)   
  QuadList.append(2*quadResult[0]/pi)
  SmallAngleList.append(2*smallAngle/pi)
  SeriesList.append(2*seriesval/pi)
```

```
## Visualization
title("Compare integration results (normalized)")
xlabel("$\\theta$")
ylabel("K(m)/K(0)")


plot(thetdeg, SimpsonList, 'b-', label = "Simpson Rule")
plot(thetdeg, QuadList, 'r-', label='Clenshaw–Curtis')
plot(thetdeg, SmallAngleList, 'g-', label='small angle' )
plot(thetdeg, SeriesList, 'm-', label='Series' )



legend(loc=2)
```

Varying $n$ in series 

Series

```
## We obtain values for differents values of 'n' (numbers)

numbers = [5,10,100,1000,2000]
SeriesComList = np.zeros((5,5))
IntervalT = np.zeros((5,5))

for i in range(0,5):
  for k in range(0,5):
    startt = time.time()
    SeriesResult = series(numbers[i],1/sqrt(2)) # Aproximation for series to integral ()
    endt = time.time()
    Intervaltime = endt-startt 
    SeriesValuetExp=sum(SeriesResult) 
    IntervalT[i,k] = Intervaltime
    SeriesComList[i,k] = SeriesValuetExp/sqrt(2)
## 
## A = [n for row, time for colum] 
```

```
# We average over the total elements
SeriesTimeMean = []
SeriesValueMean = []
for j in range(0,5):
  SeriesTimeMean.append(sum(IntervalT[j,:])/5)
  SeriesValueMean.append(sum(SeriesComList[j,:])/5)
```

```
Values = {'Number': ['5','10','100','1000','2000'],
        'Series Mean Runtime': [SeriesTimeMean[0],SeriesTimeMean[1],SeriesTimeMean[2],SeriesTimeMean[3],SeriesTimeMean[4]],
        'Series Mean Value': [SeriesValueMean[0],SeriesValueMean[1],SeriesValueMean[2],SeriesValueMean[3],SeriesValueMean[4]]
        }
df = pd.DataFrame(Values, columns = ['Number','Series Mean Value' ,'Series Mean Runtime'])
print (df)
```
